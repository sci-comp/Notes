Undergraduate
-------------

CS

1. `Introduction to Programming` (3 credits)
    1. **An Introduction to R - CRAN** by *Venables, Smith, and team* [~](https://cran.r-project.org/doc/manuals/R-intro.pdf) 
    2. **The R Inferno** by *Patrick Burns*

Math

1. `Elementary Linear Algebra` (3 credits)
    1. **Linear Algebra** by *Friedberg, Insel, and Spence* [~](http://math.stackexchange.com/questions/804716/very-good-linear-algebra-book)
    
2. `Calculus` (4 credits)
    1. **Calculus** by *Spivak* [~](http://math.stackexchange.com/questions/322892/calculus-book-recommendations-for-complete-beginner)
    
3. `Vector Calculus` (3 credits)
    1. **Calculus On Manifolds: A Modern Approach To Classical Theorems Of Advanced Calculus** by *Spivak* [~](http://math.stackexchange.com/questions/43290/references-for-the-multivariable-calculus?lq=1)
    
4. `Multivariable Calculus` (3 credits)
    1. **Multidimensional Real Analysis I and II** by *Duistermaat and Kolk* [~](http://math.stackexchange.com/questions/44522/theoretical-multivariable-calculus-textbooks?lq=1)
    
5. `Real Analysis` (3 credits)
    1. **Principles of Mathematical Analysis** by *Rudin* [~](http://math.stackexchange.com/questions/62212/good-book-for-self-study-of-a-first-course-in-real-analysis)
   
    

Stats 

1. `Statistical Methods` (6 credits)
    1. **Statistical Methods** by *Freund, Mohr, and Wilson*

2. `Theoretical Statistics` (9 credits)
    1. **Introduction to Probability and Mathematical Statistics** by *Bain and Englehardt*
    
3. `Methods of Regression Analysis` (3 credits)
    1. **Data Analysis Using Regression and Multilevel/Hierarchical Models** by *Gelman and Hill* [~](http://stats.stackexchange.com/questions/48473/recommend-a-graduate-level-regression-textbook)
    
4. `Experimental Designs` (3 credits)
    1. **Statistics for Experimenters: Design, Innovation, and Discovery** by *Box, J. S. Hunter, and W. G. Hunter*

Graduate
--------

1. `Inference Fundamentals` (6 credits)
    1. **Statistical Inference** by *Casella and Berger* [~](http://math.stackexchange.com/questions/1312087/reference-request-statistical-inference?rq=1)
    2. **Theory of Point Estimation** by *Lehmann and Casella*
    3. **Testing Statistical Hypotheses** by *Romano and Casella*

2. `Regression and ANOVA` (3 credits)
    1. **Practical Regression and Anova with R** by *Faraway* [~](https://cran.r-project.org/doc/contrib/Faraway-PRA.pdf)

3. `Probability and Distribution Theory` (3 credits)
    1. **A Probability Path** by *Resnick*
    
4. `Linear Models Theory` (3 credits)
    1. **Linear Regression Analysis** by *Seber and Lee*

5. `Experimental Design and Analysis` (3 credits)
    1. **A First Course in Design and Analysis of Experiments** by *Oehlert*

Computational Track
-------------------

1. `Statistical Computing` (3 credits)
    1. **Data Smart: Using Data Science to Transform Information into Insight** by *Foreman*
    
2. `Advanced Monte Carlo Methods`
    1. **Monte Carlo Statistical Methods** by *Robert and Casella* [~](http://stats.stackexchange.com/questions/5885/good-sources-for-learning-markov-chain-monte-carlo-mcmc?lq=1)
    
3. `Bayesian Statistics` (3 credits)
    1. **Bayesian Data Analysis** by *Gelman, Carlin, Stern, and Rudin* [~](http://stats.stackexchange.com/questions/125/what-is-the-best-introductory-bayesian-statistics-textbook)
    
4. `Data Analytics` (6 credits)
    1. **An Introduction to Statistical Learning: with Applications in R** by *James, Witten, Hastie, and Tibshirani*
    2. **The Elements of Statistical Learning: Data Mining, Inference, and Prediction** by *Hastie, Tibshirani, and Friedman* [~](http://stats.stackexchange.com/questions/33197/advanced-statistics-books-recommendation)
    
5. `Stochastic Modeling and Inference` (3 credits)
    1. **Introduction to Stochastic Processes** by *Hoel, Port, and Stone*
